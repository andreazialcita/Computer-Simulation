```python
## Transition probability matrix
A = np.array([[0.99, 0.12],[0.01, 0.88]])
A
```

```python
## Transition matrix
A_19 = np.linalg.matrix_power(A, 19)
A_19
```

```python
## Initial state distribution, s_0
s_0 = np.array([[0],[1]])
s_0
```

```python
## State of the 20th item
s_20 = np.matmul(A_19,s_0)

print('The probability that the 20th item is defective: ' + str(s_20[1,0]))
```

```python
from fractions import Fraction
```

```python
## Transition probability matrix
B = np.array([[0.5, 0.2, 0.1],[0.4, 0.5, 0.2],[0.1, 0.3, 0.7]])
B
```

```python
## Steady state distribution

B_l = np.linalg.matrix_power(B, 100)
S = Fraction(B_l[2,0]).limit_denominator()

print('Fraction of stops that can be expected to be late: ' + str(S))
```
