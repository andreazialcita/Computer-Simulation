```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

**Simulate and Evaluate**

```python
# Definitions

t_end = 21
dt = 0.005   # in hours
x_axis = np.arange(0,t_end,dt)

r = 0.1      # growth_rate
num_data_points = int(t_end/dt)
population = np.zeros(num_data_points)
population[0] = 100
```

```python
for i in range(num_data_points-1):
  growth = population[i] * r
  population[i+1] = population[i] + growth * dt
```

**Population at t = 0.050h and t = 20.000h to six decimal places**

```python
a = '%.6f'%population[10]    # population at t=0.050
b = '%.6f'%population[4000]  # population at t=20
print('Population \n t=0.050 hours: ' + a + '\n t=20 hours: ' + b )
```

**Estimated time vs population graph**

```python
plt.plot(x_axis, population)
plt.xlabel('Time (hours)')
plt.ylabel('Population')
plt.title('Population Growth Model')
```

**Analysis of the trend**

Based on the graph, the population growth seems to be exponential, meaning that the population is increasing at an increasing rate. This can be deduced from the shape of the graph. It becomes steeper as t increases â€” that is, if we draw lines tangent to the graph at different hours, we see how the tangent lines on the right of the graph (greater values of t) will be steeper.
